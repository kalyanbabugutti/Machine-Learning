# Multiple Linear Regression

Set up

```python
import warnings 
import seaborn as sns
import numpy as np  
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn import (datasets, linear_model, metrics, model_selection)
warnings.filterwarnings("ignore")   
```

```python
# Loading the Dataset
home_prices = pd.read_csv(r'homeprices_edited.csv')

# Checking the First 5 Rows of the Dataset
print(home_prices.head())
```

```python
# Chceking the information of the dataset
print(home_prices.info())
```

```python
home_prices['bedrooms'] = home_prices['bedrooms'].fillna(home_prices['bedrooms'].mean())
```

```python
# Checeking the information of the dataset
home_prices.info()
```

```python
# Separating the Features and Target Variable
x = home_prices[['age','bedrooms','area']]
y = home_prices['price']
```

```python
# Buliding the Model 
model = linear_model.LinearRegression() 
model.fit(x, y)

model.predict(x)

# Checking the efficeny of the model
print(model.score(x, y))
```

```python
# Visulization of the Data usiong the Seaborn 
sns.regplot(x='area',y='price',data=home_prices,ci=None,line_kws={'color':'red'}) 
```
