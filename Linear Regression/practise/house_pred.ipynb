```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

# Predicting the House Prices

A Real state agent collects the following data on the house prices ans sizes
1. What is the predicted price for a house with 1800 sq ft?
2. If the house size increases by 1 sq ft, how much does the price increase?
3. What is the intercept in this equation, and what does it represent?

```python
# importing the dataset
house_prices = pd.read_csv("canada_per_capita_income (1).csv")

# checking the first 5 rows of the dataset
print(house_prices.head())
```

```python
# Checking the Info of the dataset
house_prices.info()
```

# Visulation of the Graph

```python
sns.pairplot(x_vars=['year'], y_vars=['per capita income (US$)'], data=house_prices, height=5, aspect=1.5,kind='scatter')
plt.title('House Prediction')
```

```python
# Splitting the Data
x = house_prices[['year']]
y = house_prices['per capita income (US$)']
```

```python
# Buliding the Model
import sklearn
model = sklearn.linear_model.LinearRegression()
model.fit(x, y)
```

```python
# Checking the Accuarcy of the Model
model.score(x, y)
```

```python
# Drawing the Regression Line
sns.regplot(x='year', y='per capita income (US$)', data=house_prices, ci=None ,line_kws={'color': 'red'})
```
